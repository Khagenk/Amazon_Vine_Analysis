```python
# Import dependencies
import pandas as pd

# File to load
vine_data_to_load = "./vine_table.csv"

# Read vine data
vine_df = pd.read_csv(vine_data_to_load)
vine_df
```

```python
 # Filter the data and create a new DF to retrieve all rows where total_votes >= 20
helpful_votes_df = vine_df[vine_df['total_votes'] >= 20]
helpful_votes_df
```

```python
# Create new DF to retrieve all rows where helpful_votes/total_votes >= 50%
# Check data types
helpful_votes_df.dtypes
```

```python
Better_helpful_df = helpful_votes_df[helpful_votes_df['helpful_votes'] / helpful_votes_df['total_votes'] >= .50]
Better_helpful_df
```

```python
# Create new DF to retrieve all rows where review was part of Vine program (paid)
vine_paid_df = Better_helpful_df[Better_helpful_df.vine == 'Y']
vine_paid_df
```

```python
# Create new DF to retrieve all rows where review was not part of Vine (unpaid)
vine_unpaid_df = Better_helpful_df[Better_helpful_df.vine == 'N']
vine_unpaid_df
```

```python
# Determine total number of reviews, number of 5-star reviews, and percentage of 5-star reviews for paid and unpaid reviews
# Total number of reviews
len(vine_df)
```

```python
# Total number of 5-star reviews
len(vine_df[vine_df['star_rating'] == 5])
```

```python
# Total number of Vine (paid) 5-star reviews
len(vine_paid_df[vine_paid_df['star_rating'] == 5])
```

```python
# Percentage of Vine (paid) 5-star reviews
len(vine_paid_df[vine_paid_df['star_rating'] == 5]) / len(vine_paid_df) * 100
```

```python
# Total number of unpaid 5-star reviews
len(vine_unpaid_df[vine_unpaid_df['star_rating'] == 5])
```

```python
# Percentage of unpaid 5-star reviews
len(vine_unpaid_df[vine_unpaid_df['star_rating'] == 5]) / len(vine_unpaid_df) * 100
```
